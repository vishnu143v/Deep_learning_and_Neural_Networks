```python
import numpy as np
```

```python
from sklearn.linear_model import Perceptron
```

# AND gate

```python
# features X1, X2
X = np.array([[0,0],[1,0],[0,1],[1,1]])

#expected values
y_exp = np.array([0,0,0,1])

print("expected output", y_exp)

# Perceptron object is being created

clf = Perceptron()

# The model is being trainned by the data
clf.fit(X,y_exp)

#predicting the results

results = clf.predict(X)

# printing the results
print('predicted output',results)
```

#data visualization

```python
import matplotlib.pyplot as plt
```

```python
x1_min, x2_min = X[:,0].min()-1, X[:,1].min()-1
x1_max, x2_max = X[:,0].max()+1, X[:,1].max()+1

xx1 , xx2 = np.meshgrid(np.arange(x1_min,x1_max,0.1), np.arange(x2_min,x2_max,0.1))

z = clf.predict(np.c_[xx1.ravel(),xx2.ravel()])
z=z.reshape(xx1.shape)

plt.contourf(xx1,xx2,z,alpha=0.2)

plt.scatter(X[:,0],X[:,1], c= y_exp ,cmap = plt.cm.brg)

plt.xlabel('Feature 1')

plt.ylabel('Feature 2')

plt.xlim(xx1.min(),xx1.max())

plt.ylim(xx2.min(),xx2.max())

plt.show()
```
